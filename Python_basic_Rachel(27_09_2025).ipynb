# Conditional Statements

* To make decision and excute particulae section of code
* Keyword: If, else,elif

```python
# if: When we have only one condition to check
age=15
if age>=18:
  print('They can vote')
```

```python
# else: When we have two condition to check
age=20
if age>=18:
  print('They can vote')
else:
  print('you are not eligible')
```

```python
# elif: When we have multiple conditions to check
age=23
if age>=13 and age<=19:
  print('ypu are a teenager')
elif age==30:
  print('you are in your 30')
elif age>=20:
  print('You are an adult')
else:
  print('You are a  minor')
```

```python
color='Pink'

if color=='red':
  print('color of blood')
elif color=='Blue':
  print('color of calm')
else:
  print('some other color')
```

# While loop

# Nested Loop

* Loop inside loop

```python
for i in range(2,7):
  for j in range(1,4):
    print(f'{i} and {j}')
```

```python
for i in range(1,11):
  for j in range(1,11):
    print(f'{i} X {j}= {i*j}')
    print()
  print()
```

```python
for row in range(3):
  for col in range(5):
    print('*',end=' ')
  print()
```

```python
for i in range(1,6):
  for j in range(i):
    print('*',end=' ')
  print()
```

```python
for i in range(5,0,-1):
  for j in range(i):
    print('*',end=' ')
  print()
```

```python
for i in range(1,6):
  for j in range(6-i):
    print(' ',end='')

  for k in range(2*i-1):
    print('*', end='')
  print()
```

```python
for i in range(1,6):
  for j in range(6-i):
    print('*',end=' ')
  print()
```
