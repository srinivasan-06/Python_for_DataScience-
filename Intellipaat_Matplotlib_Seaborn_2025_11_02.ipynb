# <mark>Scipy

Find the Roots of a Quadratic Equation
+ Goal: Solve a quadratic equation




2
+


+

=
0
using scipy.optimize.root.

```python
from scipy.optimize import root
```

```python
def quadratic_equation(x):
  # equation x^2 - 3x + 2 = 0
  a, b, c = 1, -3, 2
  return (a * x**2) + (b*x) + c


initial_guess = [0]
solution = root(quadratic_equation, initial_guess)
print(f'Root: {solution.x}')
```

Compute the Integral of a Function
Goal: Compute the definite integral of


()=^2


 from 0 to 2 using scipy.integrate.quad.

```python
from scipy.integrate import quad
```

```python
def f(x):
  return x**2

result, error = quad(f,0,2)
print(f'Integral result: {result}')
```

 Solve a System of Linear Equations

Goal: Solve a system of equations using scipy.linalg.solve.


Example system:

2+3=8



4=2


```python
from scipy.linalg import solve
```

```python
# define coefficients and constants
A = [[2,3], [1,-4]]   # coefficients
B = [8, -2]           # constants

solution = solve(A,B)
print(f'Solution: {solution}')
```
