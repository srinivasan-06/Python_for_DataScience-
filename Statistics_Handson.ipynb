```python
plt.figure(figsize =(15, 12))
sns.barplot(x= 'country',
            y ='Economy: Services and other activity (% of GVA)',
            data = df)


plt.xticks(rotation = 90)
plt.title('service sector contribution by country')
plt.ylabel('% of economy')
plt.show()
```

```python
# distribution of population in thousand
population = df['Population in thousands (2017)']
sns.histplot(population, bins =10, kde = True , color ='red')
plt.title('popuation distrinution with kde curve')
plt.xlabel('population')
plt.ylabel('no of countries')
plt.grid()
plt.show()
```
