```python
# range : it tells us  how spread out the data is


# difference between the highest and lowest number in dataset


# if range is small , you data points are close

# if range is largest ,  your data are spread out
```

```python
arr = [1,2,3,4,5,20,40,50,60,100,200,300,400]
# range - spread

maximum = max(arr)

minimum = min(arr)


range = maximum  - minimum

print("Maximum = {}, Minimum = {} and Range = {}" .format(maximum, minimum , range))





```

```python
# 2. variance:
# variance tells how spread out the data is

# if variance is small -> the number are close to average
# if variance is large - the number are spread out



from statistics import variance

data =[2,4,6,10,12]

print("varaince", variance(data))


# 17.2 -> spread out or scattered from each other


# machine lerning : average square of spread in dataset


```

```python
# standard deviation : standard deviation  is measure that shows how spread

# out the number are from their average . it is calculate by finding out the
# average square dfference from the mean and then taking the square root of that average


from statistics import stdev

data = [5,6,7,8,9,10,23,4,5,6,7]

result = stdev(data)
print("standard deviation is", result)
```

```python
# difference between varaince and standard deviation


# 1. variance : average of squared distance from the mean
# value is usually bigger

# 2. standarddeviation : square root of variance

# value is smaller
```

```python
# Percentile :


# suppose 100students took a test

# you score 85 marks
# how many students scored less than me..

#  90th percentile : there are 90 student who score less than you

# percentile will tell you: your position as compared to others




data = [ 50,60,70,80,90] # sample score


#find 75th percentile

p75 = np.percentile(data , 75)

print(p75)

# 25th percentile

p25 = np.percentile(data, 25)
print(p25)

# 50th percentile -  median

p50 = np.percentile(data, 50)
print(p50)





# this means 75% of the score are below 80
# 25% of the scores are below 60

# sales

# percentile : 25, 50, 75:  65, 75,  85



# 10,000 - 25th


#  low, average and high

# percentile tell us how the score are spread out and help us understand who scored
# low, average and high
```

```python
# quartiles:


# imagine we want to sort the number from small to big and then cut them into 4 equal groupd

# are called quartile

# there are 3 quartile


# Q1 (FIRST Quartile )

#  the value below which 25% of the data lies



#Q2(second quartile / median) :

#  the value below 50% of the data lies
# half of your data


# q3(Third quartile):
# the value below which 75% of the data lies




# example: suppose you have test score of 12 scores

# 40, 45,50,55,60,62,65, 70,72, 80, 85, 90

# group 1:  40 , 45, 50  Q1

# group 2:  55, 60, 62 -> q2 (median)
# group 3: 65, 70,72  #q3

# group 4: 80, 85, 90



```

```python
data = [40, 45,50,55,60,62,65, 70,72, 80, 85, 90]
import pandas as pd
s= pd.Series(data)



q1 = s.quantile(0.25)
q2  = s.quantile(0.50)
q3 = s.quantile(0.75)



print(q1, q2, q3)

# 25% of my data values are below  53.75

# 50%of the my data values are below 63.5

# 75% of the my data values are below 74


```

```python
# outliers :
age = [20,30,40,50,60,70,10,23,43,12,34,56,78,98,100,200,-250,271]

# detect these outliers

import matplotlib.pyplot as plt
plt.boxplot(age)
plt.show()

# some oultier are there
```

```python
age = [20,30,40,50,60,70,10,23,43,12,34,56,78,98,100,200,-250,271]

quartile_1 = np.percentile(age , 25)
quartile_3 = np.percentile(age, 75)

print("Q1", quartile_1)
print("Q3", quartile_3)


# middle range = 50 percentile

Interquartile_range = quartile_3 - quartile_1
print("Interquartile range", Interquartile_range)


```

```python
# lets remove the outliers:

upper_fence =  quartile_3 + 1.5 *Interquartile_range

lower_fence =  quartile_1 - 1.5* Interquartile_range


print(upper_fence, lower_fence)


# 152  - will be removed

 # - 52 will be removed
```

```python
for  i in age:
  if i >=upper_fence or i <=lower_fence:
    print("outlier value in data", i, " and it been deleted")
    age.remove(i)
print(age)
```

**Kurtosis**

kurtosis is a way to descrive the shape of a data distribution

especially how tall or flat the peak of the graph


it tells us how the values are spread , wether they are tightly packed around the centre or more spread out



**type of kurtosis**


1. leptokurtic:(high peak)

the graph has a very sharp tall peak

2. Mesokurtic(Normal peak)

when the graph has a moderate peak , like normal bell curve
data is evenly spread out around the centree



3. PlatyKurtic(Flat peak)

the graph has a low , flat peak

data is more spread out , not tightly packed in the centre

**Skewness**


skewness tells us about the symmetry of our data

if a distribution is prefectly symmetric - like bell shape - skewness = 0


**types of skewness**:

1. Positive skewness (right skewed)


the tail on ride side (higher values) is longer or fatter

most data points are concentrated on the left side

skewness value is greater than 0



2. negative skewness (left skewed):

the tail on the  left side is longer or fatter


most data is concentrated on the right


skewness value is less than 0


3. zero skewness (symmetric)

the distribution is balanced on both side

skewness = 0
