# OOP: Object Oriented Programming

1. Class : Container to store variable(attributes) and function(methods)
2. Object: Used to access the variable and fucntion stored inside the class.

```python
class Data:
  name='Rizwan'
  state='UP'
  marks=90

  def display(self):
    print(self.name,self.state,self.marks)

obj1=Data()
obj1.display()
obj1.marks
```

# Constructors(init):
Constructors is a special method in a class used to initialize the value of the object variables when the object is created.

```python
class Data:
  def __init__(self,name,state,marks):
    self.name=name
    self.state=state
    self.marks=marks

  def display(self):
    print(self.name,self.state,self.marks)

obj1=Data('Ananya','MP',100)
obj1.display()
obj2=Data('faizan','Rajasthan',98)
obj2.display()
```

```python
class Account:
  def __init__(self,holder):
    self.holder=holder

  def show(self):
    print('account holder name is',self.holder)


obj1=Account('Rachel')
obj1.show()
obj2=Account('Ananya')
obj2.show()
```

# Four Pillars of OOPs

1. Inheritance: Let one class reuse code from another.
2. Polymorsiam: One name , many form
3. Encapsulation: Hide internal details, only expose what's needed.
4. Abstraction: show only important things, hide unnecessary details.

```python
# Inheritance -->inheritance variable and fuctions from the parent class.
class Animal:
  def display(self):
    print('This is the animal class')

class Cat(Animal):
  def show(self):
    print('This is the child class')

obj1=Cat()
obj1.display()
obj1.show()
```

# 1. Single Inheritance

one parent-->one child

```python
class Animal:
  def display(self):
    print('This is the animal class')

class Cat(Animal):
  def show(self):
    print('This is the child class')

obj1=Cat()
obj1.display()
obj1.show()
```

# 2. Hierarchical Inheritance

one parent --> child 1, child 2, child 3.....

```python
class Animal:
  def display(self):
    print('animal')

class Dog(Animal):
  def show(self):
    print('Dog')

class Cat(Animal):
  def view(self):
    print('Cat')


obj1=Cat()
obj1.display()
obj1.view()

obj2=Dog()
obj2.show()
obj2.display()
```

# Multiple Inheritance:

Parent 1, parent 2, parent 3....-->Child 1

```python
class Animal:
  def display(self):
    print('Animal')

class Mammel:
  def View(self):
    print('Mammel')

class Dog(Animal,Mammel):
  def show(self):
    print('dog')

obj_dog=Dog()
obj_dog.View()
obj_dog.display()
obj_dog.show()
```

```python
# case 1:
class Animal:
  def display(self):
    print('Animal')

class Mammel:
  def View(self):
    print('Mammel')

class Dog(Animal,Mammel):
  def display(self):
    print('dog')

obj_dog=Dog()
obj_dog.display()
```

```python
# case 2:
class Animal:
  def display(self):
    print('Animal')

class Mammel:
  def display(self):
    print('Mammel')

class Dog(Animal,Mammel):
  def show(self):
    print('dog')

obj_dog=Dog()
obj_dog.display()
```

```python
# Case 3:
class Animal:
  def display(self):
    print('Animal')

class Mammel:
  def View(self):
    print('Mammel')

class Dog(Animal,Mammel):
  def display(self):
    print('dog')
    super().display()

obj_dog=Dog()

obj_dog.display()
```

```python
# case 4:
class Animal:
  def display(self):
    print('Animal')

class Mammel:
  def display(self):
    print('Mammel')

class Dog(Animal,Mammel):
  def display(self):
    print('dog')
    Animal.display(self)
    Mammel.display(self)

obj_dog=Dog()
obj_dog.display()
```

# Multilevel Inheritance:

grandparent--->parent--->son

```python
class Grandparent:
  def display(self):
    print('This is the grand parent class')

class Parent(Grandparent):
  def show(self):
    print('this is the parent class')

class Son(Parent):
  def view(self):
    print('This is the son class')


obj1=Son()
obj1.show()
obj1.display()
obj1.view()
```

# Polymorphism:

Same name of the methods will be used in the different classes according to the requirement.

```python
class Square:
  def __init__(self,side):
    self.side=side

  def area(self):
    print(self.side*self.side)

class Rectangle:
  def __init__(self,length, breadth):
    self.length=length
    self.breadth=breadth

  def area(self):
    print(self.length*self.breadth)

class Traingle:
  def __init__(self,base,height):
    self.base=base
    self.height=height

  def area(self):
    print(0.5*self.base*self.height)



obj1=Traingle(3,9)
obj1.area()

obj2=Rectangle(3,9)
obj2.area()

obj3=Square(2)
obj3.area()
```

```python
# Method Overloading

class mathmatical:
  def add(self,a,b):
    print(a+b)

  def add(self,a,b,c):
    print(a+b+c)

obj1=mathmatical()
obj1.add(2,3,4)
```

```python
# method overriding

class Bird:
  def fly(self):
    print('the bird is flying')

class Airplane(Bird):
  def fly(self):
    super().fly()
    print('This airplane is flying')

obj1=Airplane()
obj1.fly()
```

# Encapsulation

Encapsulation means binding data(varible) and code(methods) togehter in a class and restricting direct access to some of the object's.


```python
class Details:
  name='pooja'# public
  _adharnum='123abc'# protected
  __password='@123'# private


obj1=Details()
print(obj1.name)
print(obj1._adharnum)
# obj1.__password
```

```python
class Details:
  name='pooja'# public
  _adharnum='123abc'# protected
  __password='@123'# private

  def display(self):
    print(self.__password)


obj1=Details()
obj1.display()
```

```python
class Details:
  def __init__(self):
    self.__password='@123'

  def __show(self):
    return self.__password

  def _access(self):
    return self.__show()

obj1=Details()
obj1._access()
```

# Abstraction

Abstraction means hiding the complex internal details and only the essesntial features to the user.

```python
from abc import ABC, abstractmethod

class abcclass(ABC):
  @abstractmethod
  def display(self):
    print('complex details')


obj1=abcclass()
obj1.display()
```

```python
from abc import ABC, abstractmethod

class abcclass(ABC):
  @abstractmethod
  def display(self):
    print('complex details')

class dummy(abcclass):
  def display(self):
    super().display()


obj1=dummy()
obj1.display()
```

```python
from abc import ABC, abstractmethod

class Phone(ABC):
  @abstractmethod
  def call(self):
    print('Number are getting converted into packet')

  def charging(self):
    print('all phones support charging')

class dummy(Phone):
  def call(self):
    super().call()


obj1=dummy()
obj1.call()
obj1.charging()
```
