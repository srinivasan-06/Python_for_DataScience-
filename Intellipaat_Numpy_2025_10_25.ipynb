Element wise Addition

```python
maths =   [23,46,57,80]
science = [47,68,23,76]
```

```python
#  total score of student 1 using python way
s1_score = maths[0] + science[0]
s1_score
```

```python
scores = []

for score_1, score_2 in zip(maths, science):
  score = score_1 + score_2
  scores.append(score)

scores
```

```python
#   sum of individual elements using numpy way
np.sum([maths, science])
```

```python
#  sum of individual student scores
np.sum([maths, science], axis=0)
```

```python
#  sum of individual subject scores
np.sum([maths, science], axis=1)
```
