```python
import time
```

```python
a = [1,2,3]
b = [4,5,6]
```

```python
#  python loop
for i in range(len(a)):
  print(a[i], '+', b[i], ' = ', a[i]+b[i])
```

```python
# list comprehension
[a[i]+b[i] for i in range(len(a))]
```

```python
# using python
def using_list(n):
  # start time
  start_time = time.time()

  list_1 = range(n)
  list_2 = range(n)

  # element wise addition using python
  result = [list_1[i] + list_2[i]  for i in range(len(list_1))]

  # end time
  end_time = time.time()

  # duration
  time_diff = end_time - start_time

  return time_diff
```

```python
# using numpy
def using_array(n):
  # start time
  start_time = time.time()

  array_1 = np.arange(n)
  array_2 = np.arange(n)

  # element wise addition using numpy
  result = array_1 + array_2

  # end time
  end_time = time.time()

  # duration
  time_diff = end_time - start_time

  return time_diff
```

```python
using_list(1000000)
```

```python
using_array(1000000)
```

```python
print(f'Numpy is {round(0.21904253959655762/0.016629457473754883, 0)} times faster than python')
```

```python
using_list(100000000)
```

```python
using_array(100000000)
```

```python
print(f'Numpy is {round(24.215423822402954/1.5421640872955322, 0)} times faster than python')
```
